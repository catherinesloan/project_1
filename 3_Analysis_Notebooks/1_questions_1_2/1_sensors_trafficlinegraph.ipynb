```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
# Include this line to make plots interactive
%matplotlib notebook
```

```python
# pulling the data from the csv file
df = pd.read_csv("../../Output/1_pedestrian_csv/censors2019_chosen3_dailyaverage.csv")
df.head()
```

```python
# get total number of data entries
df.count()
```

```python
# confirm data types
df.dtypes
```

```python
# set x-axis for line plot
x_axis = np.arange(1,363,1)
```

```python
# set y-axis data as variables to plot each sensor location
points_b = df["Bourke Street Mall (North)"]
points_f = df["Flinders St-Spark La"]
points_q = df["QV Market-Elizabeth St (West)"]
```

```python
# plot each sensor location as a line plot
bourke, = plt.plot(x_axis, points_b, color="blue", linewidth=0.5, label="Bourke Street Mall (North)")
flinders, = plt.plot(x_axis, points_f, color="Red", linewidth=0.5, label="Flinders St-Spark La")
qvm, = plt.plot(x_axis, points_q, color="green", linewidth=0.5, label="QV Market-Elizabeth St (West)")

# Create labels for the X and Y axis
plt.xlabel("Days in a year")
plt.ylabel("Average daily foot traffic")
plt.title("Comparing average daily foot traffic in chosen 3 locations")

# Set our legend to where the chart thinks is best
plt.legend(handles=[bourke, flinders, qvm], loc="best")

plt.savefig("../../Output/linegraph_3locations_dailyavfoottraffic.png")

plt.show()
```
