```python
# Dependencies and Setup
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import requests
import pandas as pd
import json
from citipy import citipy
import random

# Import API key
from config import api_key
```

```python
# Base URL required for API call
url = "https://developers.zomato.com/api/v2.1/search?"

# Parameters to run through Query URL
start = np.arange(0,20,20)
count = 20
city = 259
lat = -37.8136
lon = 144.9631
radius = 2500

# Empty list to hold all API calls
response_data = []

# Loop through Zomato API
for a in range(len(start)):
    query_url = f"{url}entity_id={city}&entity_type=metro&lat={lat}&lon={lon}&radius={radius}&sort=real_distance&apikey={api_key}&start={start[a]}&count={count}"
    response_data.append(requests.get(query_url).json())
```

```python
# Empty list to hold all restaurant names, aggregate rating, hours of operation, type of eatery, price 
restr_names = []
rating = []
cuisine_type = []
price_range = []
restr_lat = []
restr_lon = []
b = 0

for a in range(len(start)):
    for x in range(count):
        restr_names.append(response_data[b]["restaurants"][x]["restaurant"]["name"])
        rating.append(response_data[b]["restaurants"][x]["restaurant"]["user_rating"]["aggregate_rating"])
        cuisine_type.append(response_data[b]["restaurants"][x]["restaurant"]["cuisines"])
        price_range.append(response_data[b]["restaurants"][x]["restaurant"]["price_range"])
        restr_lat.append(response_data[b]["restaurants"][x]["restaurant"]["location"]["latitude"])
        restr_lon.append(response_data[b]["restaurants"][x]["restaurant"]["location"]["longitude"])        
    b += 1
```

```python
zomato_df = pd.DataFrame({
    "Name": restr_names,
    "Aggregate Rating": rating,
    "Type": cuisine_type,
    "Price": price_range,
    "Latitude": restr_lat,
    "Longitude": restr_lon,
})

zomato_df
```
