```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import json
```

```python
# Pull the data from the csv
ped_count_df = pd.read_csv("Resource/COM_ftraffic_2019_select4.csv")
# Change the -ve values to 0
ped_num_df = ped_count_df._get_numeric_data()
ped_num_df[ped_num_df < 0] = 0
# Convert the date into a datetime for easy sorting
ped_count_df['Date'] = pd.to_datetime(ped_count_df['Date'])
ped_count_df
```

```python
# Get the daily average pedestrian count
averagePed_df = ped_count_df.groupby('Date')[['Bourke Street Mall (North)', 'QV Market-Peel St', 'Flinders St-Spring St (West)', 'QVM-Franklin St (North)']].mean()
averagePed_df
```

```python
# Transpose the data ready to merge to the Zomato data
averagePed_df_transposed = averagePed_df.T
averagePed_df_transposed
```
